**Условие 2: 2 задача**

Постройте график

2.2. Изучите распределение этажей домов
2.3. Изучите распределение состояния домов



```python
data = df['view'].value_counts()
data.index = [0, 1, 2, 3, 4]
data.head()
```

```python
plt.figure(figsize=(16, 4))

plt.pie(data, autopct='%1.1f%%')

plt.title('Доля домов, имеющих вид на набережную по количеству окон')
plt.legend(data.index);
```

```python
data = df['floors'].value_counts()
data.index = [0, 1, 2, 3, 3,5]
data.head()
```

```python
plt.figure(figsize=(16, 4))

plt.pie(data, labels = data.index)

plt.title('Распределение домов по колличеству этажей')
plt.legend(data.index);
```

```python
data = df['condition'].value_counts()
data.index = [1, 2, 3, 4, 5]
data.head()
```

```python
plt.figure(figsize=(16, 4))
explode=[0, 0.1, 0.2, 0.3, 0.4]

plt.pie(data, explode, labels = data.index)

plt.title('Распределение домов по состоянию')
plt.legend(data.index);
```
